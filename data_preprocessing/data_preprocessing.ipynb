# I. Dataframe

## 1. Dataframe thứ nhất

```python
import pandas as pd

# Đọc dữ liệu từ file CSV
data = pd.read_csv('../airbnb-usa-ny-newyork/20231204-listings-detailed.csv')

data.head()
```

```python
# Select only the features relevant for predicting house prices on Airbnb without reviews
relevant_features = [
    "id", "listing_url", "name", "picture_url", "latitude", "longitude", "property_type", "room_type", "accommodates",
    "bathrooms_text", "bedrooms", "beds", "amenities", "price", "minimum_nights", "maximum_nights",
    "availability_30", "availability_365"
]

# Filter the dataframe with relevant features
filtered_df = data[relevant_features]
```

## 2. Lấy dataframe thứ hai và gộp với dataframe thứ nhất

```python
# Load the second dataframe
df_neighbourhood = pd.read_csv('/kaggle/input/airbnb-usa-ny-newyork/20231204-listings-summary.csv')

df_neighbourhood_selected = df_neighbourhood[["id", "neighbourhood_group", "neighbourhood"]]

# Merge the two dataframes on "id"
airbnb_data = pd.merge(filtered_df, df_neighbourhood_selected, on="id", how="left")
```

## 3. Lấy ra mẫu nhỏ hơn để tính toán

# II. Xử lí giá trị rỗng

```python
airbnb_data.isnull().sum()
```

## 1. Tạo giá trị bedrooms từ cột name

```python
import re

# Function to extract the number of bedrooms from the "name" column
def extract_bedrooms(name):
    match = re.search(r'(\d+)\s+bedrooms?', name)
    if match:
        return int(match.group(1))
    else:
        return None
```

```python
# Apply the function to extract the number of bedrooms
airbnb_data['bedrooms'] = airbnb_data['name'].apply(extract_bedrooms)
```

```python
airbnb_data.isnull().sum()
```

## 2. Xóa các hàng thiếu giá trị quan trọng trong các cột: bedrooms, beds, bathrooms_text,...

```python
airbnb_data = airbnb_data.dropna()
```

```python
missing_data = airbnb_data.isnull().sum().sort_values(ascending=False)
print("\nDữ liệu thiếu:\n", missing_data)
```

# III. Xử lí các cột cần sửa

```python
airbnb_data.info()
```

## 1. Chuyển đổi cột 'price' từ dạng chuỗi sang dạng số


```python
airbnb_data['price'] = airbnb_data['price'].replace('[\$,]', '', regex=True).astype(float)
```

## 2. Sửa cột room_type và dùng One-Hot Encoding

```python
# Tính tần suất xuất hiện của từng loại trong room_type
room_type_freq = airbnb_data['room_type'].value_counts(normalize=True)

# Chuyển loại phòng có tần suất dưới 5% thành 'Other'
airbnb_data['room_type'] = airbnb_data['room_type'].apply(
    lambda x: 'Other' if room_type_freq[x] < 0.05 else x
)
```

```python
# Áp dụng One-Hot Encoding cho room_type
room_type_dummies = pd.get_dummies(airbnb_data['room_type'], prefix='room_type')

# Kết hợp các cột mới vào DataFrame ban đầu
airbnb_data = pd.concat([airbnb_data, room_type_dummies], axis=1)

# Xóa cột room_type nếu không cần thiết
airbnb_data = airbnb_data.drop('room_type', axis=1)
```

## 3. Lấy ra cột số phòng tắm

```python
# Lấy giá trị số đầu tiên trong cột bathrooms_text nếu có
airbnb_data['bathrooms_text'] = airbnb_data['bathrooms_text'].apply(
    lambda x: re.findall(r'\d+', str(x))[0] if pd.notnull(x) and re.findall(r'\d+', str(x)) else None
)
airbnb_data.rename(columns={'bathrooms_text': 'bathrooms'}, inplace=True)
airbnb_data['bathrooms'] = airbnb_data['bathrooms'].astype(float)
```

## 4. Sử dụng One-Hot Encoding cho cột neighbourhood_group

```python
# Sử dụng One-Hot Encoding cho cột neighbourhood_group
neighbourhood_group_dummies = pd.get_dummies(airbnb_data['neighbourhood_group'], prefix='neighbourhood_group')

# Kết hợp các cột mới vào DataFrame ban đầu
airbnb_data = pd.concat([airbnb_data, neighbourhood_group_dummies], axis=1)

# Xóa cột neighbourhood_group nếu không cần thiết
airbnb_data = airbnb_data.drop('neighbourhood_group', axis=1)
```

## 5. Xóa các cột không sử dụng

```python
airbnb_data = airbnb_data.drop(['name', 'property_type', 'id', 'neighbourhood', 'listing_url'], axis=1)
```

# IV. Thêm các cột phù hợp để dự đoán giá

## 1. Tạo cột 'distance_to_center' thay cho cột 'latitude' và 'longitude'

```python
from geopy.distance import geodesic

# Ví dụ: tạo một hàm tính khoảng cách giữa hai tọa độ
def calculate_distance(row):
    rental_location = (row['latitude'], row['longitude'])
    city_center = (40.7128, -74.0060)  # Tọa độ trung tâm của thành phố New York
    return geodesic(rental_location, city_center).kilometers

# Áp dụng hàm tính khoảng cách để tạo cột mới
airbnb_data['distance_to_center'] = airbnb_data.apply(calculate_distance, axis=1)
```

```python
airbnb_data = airbnb_data.drop(['latitude', 'longitude'], axis=1)
```

## 2. Tạo thêm cột 'beds_per_bedroom' 

```python
airbnb_data['beds_per_bedroom'] = airbnb_data['beds'] / airbnb_data['bedrooms']
```

## 3. Tạo cột 'amenity_count' từ cột 'amenities'

```python
import ast
import pandas as pd
from sklearn.preprocessing import MultiLabelBinarizer

# Chuyển đổi tiện nghi thành danh sách, thay NaN bằng danh sách rỗng
airbnb_data['amenities'] = airbnb_data['amenities'].apply(
    lambda x: ast.literal_eval(x) if isinstance(x, str) else (x if isinstance(x, list) else [])
)

# Đếm số lượng tiện nghi trong mỗi danh sách
airbnb_data['amenity_count'] = airbnb_data['amenities'].apply(len)

airbnb_data = airbnb_data.drop(['amenities'], axis=1)
```

## 4. Tạo cột chỉ số MOS: Đánh giá chất lượng bức ảnh

### 4.1. Tải ảnh của dữ liệu phòng

```python
import os
import requests
from PIL import Image
from io import BytesIO
import pandas as pd
import numpy as np

# Tạo thư mục lưu ảnh
image_dir = "images"
os.makedirs(image_dir, exist_ok=True)

def download_image(url, save_path):
    try:
        response = requests.get(url, timeout=10)
        response.raise_for_status()  # Kiểm tra lỗi HTTP
        img = Image.open(BytesIO(response.content))
        img.save(save_path)
        return True
    except Exception as e:
        return False

# Tải ảnh từ cột picture_url
airbnb_data['image_downloaded'] = airbnb_data['picture_url'].apply(
    lambda url: download_image(url, os.path.join(image_dir, f"{hash(url)}.jpg")) if pd.notnull(url) else False
)
```

### 4.2. Chỉ giữ lại những hàng tải được ảnh

```python
airbnb_data = airbnb_data[airbnb_data['image_downloaded'] != False]
```

```python
airbnb_data.shape
```

### 4.3. Khởi tạo model NIMA tính toán MOS


```python
import tensorflow as tf
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras import layers, models
import numpy as np
import cv2

# Kiểm tra các thiết bị GPU có sẵn
print("GPU Available: ", tf.config.list_physical_devices('GPU'))

gpus = tf.config.experimental.list_physical_devices('GPU')
if gpus:
    try:
        # Cấu hình để chỉ sử dụng bộ nhớ cần thiết
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
        print("Memory growth set for GPUs")
    except RuntimeError as e:
        print(e)

# Tải MobileNetV2 làm mô hình cơ sở
base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# Thêm các lớp để dự đoán điểm NIMA
x = base_model.output
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dropout(0.75)(x)
x = layers.Dense(10, activation='softmax')(x)  # 10 lớp cho các điểm từ 1 đến 10

# Tạo mô hình
model_nima = models.Model(inputs=base_model.input, outputs=x)
model_nima.compile(optimizer='adam', loss='categorical_crossentropy')

def preprocess_image(image_path):
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image = cv2.resize(image, (224, 224))  # Kích thước đầu vào cho mô hình
    image = image / 255.0  # Chuẩn hóa ảnh về khoảng [0, 1]
    image = np.expand_dims(image, axis=0)
    return image
```

### 4.3. Tính toán MOS


```python
import tensorflow as tf
from keras.applications.mobilenet import preprocess_input
from keras.preprocessing import image

def calculate_mos(img_path):
    try:
        img = image.load_img(img_path, target_size=(224, 224))
        img_array = image.img_to_array(img)
        img_array = np.expand_dims(img_array, axis=0)
        img_array = preprocess_input(img_array)

        # Dự đoán MOS
        preds = model_nima.predict(img_array)
        mos = np.dot(preds, np.arange(1, 11))  # Tính MOS từ xác suất
        return mos[0]
    except Exception as e:
        print(f"Failed to calculate MOS for {img_path}: {e}")
        return None

# Tính toán MOS cho tất cả các ảnh đã tải về
airbnb_data['mos'] = airbnb_data['picture_url'].apply(
    lambda url: calculate_mos(os.path.join(image_dir, f"{hash(url)}.jpg")) if pd.notnull(url) else None
)
```
