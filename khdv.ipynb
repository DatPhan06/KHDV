```python
import pandas as pd

# Đọc dữ liệu từ file CSV
data = pd.read_csv('airbnb-usa-ny-newyork/20231204-listings-detailed.csv')

data.head()
```

```python
data.shape
```

```python
# Select only the features relevant for predicting house prices on Airbnb without reviews
relevant_features = [
    "id", "listing_url", "name", "picture_url", "latitude", "longitude", "property_type", "room_type", "accommodates",
    "bathrooms_text", "bedrooms", "beds", "amenities", "price", "minimum_nights", "maximum_nights",
    "availability_30", "availability_365"
]

# Filter the dataframe with relevant features
filtered_df = data[relevant_features]
```

```python
filtered_df.shape
```

```python

# Load the second dataframe
df_neighbourhood = pd.read_csv('airbnb-usa-ny-newyork/20231204-listings-summary.csv')

df_neighbourhood_selected = df_neighbourhood[["id", "neighbourhood_group", "neighbourhood"]]


# Merge the two dataframes on "id"
merged_df = pd.merge(filtered_df, df_neighbourhood_selected, on="id", how="left")
```

```python
merged_df.shape
```

```python
merged_df.head()
```

```python
merged_df.isnull().sum()
```

```python
import re

# Function to extract the number of bedrooms from the "name" column
def extract_bedrooms(name):
    match = re.search(r'(\d+)\s+bedrooms?', name)
    if match:
        return int(match.group(1))
    else:
        return None
```

```python
# Apply the function to extract the number of bedrooms
merged_df['bedrooms'] = merged_df['name'].apply(extract_bedrooms)
```

```python
merged_df.isnull().sum()
```

```python
merged_df.head()
```

```python
merged_df.to_csv('full.csv', index=False)
```

```python
# 1. Loại bỏ ký tự $ trong cột `price` và chuyển về dạng float
merged_df['price'] = merged_df['price'].replace('[\$,]', '', regex=True).astype(float)
```

```python
# 2. Chuyển các cột về kiểu dữ liệu phù hợp
merged_df['minimum_nights'] = merged_df['minimum_nights'].astype(int)
merged_df['maximum_nights'] = merged_df['maximum_nights'].astype(int)
merged_df['availability_30'] = merged_df['availability_30'].astype(int)
merged_df['availability_365'] = merged_df['availability_365'].astype(int)
merged_df['bedrooms'] = merged_df['bedrooms'].astype(float)
merged_df['beds'] = merged_df['beds'].astype(float)
```

```python
# 3. Kiểm tra và thay thế giá trị null trong cột 'bathrooms_text'
merged_df['bathrooms_text'].fillna('Unknown', inplace=True)
```

```python
# 4. Xử lý cột 'amenities' (nếu cần thiết, có thể phân tích thêm chuỗi JSON)
merged_df['amenities'] = merged_df['amenities'].apply(lambda x: eval(x) if isinstance(x, str) else x)
```
